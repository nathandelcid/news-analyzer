# Trends/Averages

## Simple Moving Averages

$$\text{SMA}_N(t) = \frac{1}{N} \sum_{i=0}^{N-1}P_{t-i}$$

## Exponential Moving Average

Like SMA, but it gives more weight to recent prices 

$$\text{EMA}_t = \alpha \cdot P_t + (1-\alpha) \cdot \text{EMA}_{t-1} \\ \alpha = \frac{2}{N+1}$$

## Moving Average Convergence Divergence

MACD is built upon EMA. 
* It tells us when short-term price momentum begins to shift direction compared to long term trends

$$\text{MACD}(t) = \text{EMA}_{12}(t) - \text{EMA}_{26}(t)$$

### Signal Line

$$\text{Signal}(t) = \text{EMA}_{9}(\text{MACD}(t))$$

### Histogram

$$\text{Hist}(t) = \text{MACD}(t) - \text{Signal}(t)$$

# Volatility

## Bollinger Bands

Measures **volatility around a moving average**.
* They consist of three lines
    * **Middle Band:** usually a 20-period **SMA** (trend baseline).
    * **Upper Band:** SMA + *k × standard deviation* (often *k = 2*).
    * **Lower Band:** SMA − *k × standard deviation*.

$$\text{MB}*t = \text{SMA}*{n}(CLOSE_t)$$

$$\text{UB}_t = \text{MB}*t + k \times \text{STD}*{n}(CLOSE_t)$$

$$\text{LB}_t = \text{MB}*t - k \times \text{STD}*{n}(CLOSE_t)$$

# Momentum

## Relative Strength Index

RSI measures **momentum**, comparing the magnitude of recent gains vs losses over a window (commonly 14 periods).
* It oscillates between 0 – 100:
    * $> 70 \rightarrow$ overbought
    * $< 30 \rightarrow$ oversold

$$\text{RS} = \frac{\text{Avg Gain}*{n}}{\text{Avg Loss}*{n}}$$
$$\text{RSI} = 100 - \frac{100}{1 + \text{RS}}$$
